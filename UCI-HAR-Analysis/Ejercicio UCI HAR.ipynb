```python
import zipfile
import os
import urllib.request
```

```python
# Descargar y extraer el dataset
data_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00240/UCI%20HAR%20Dataset.zip"
zip_path = "UCI_HAR_Dataset.zip"
dataset_path =  "UCI_HAR_Dataset" # Hay que hacer que se guarde en la carpeta de los datasets

if not os.path.exists(dataset_path):
  print("Descargando el dataset...")
  urllib.request.urlretrieve(data_url, zip_path)

  with zipfile.ZipFile(zip_path, "r") as zip_ref:
    zip_ref.extractall(".") # Cambiar el contenido de las comillas donde quieras guardarlo

  print("Dataset descargado y extraido exitosamente.")
else:
  print("Dataset ya descargado.")
```
